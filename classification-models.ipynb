```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/review-1/review1.csv')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(stop_words=stop_words,use_idf=True,min_df= 0.04)
tfidf = vectorizer.fit_transform(df.text)
tfidf.shape
```

```python
df['stars'] = df['stars'].apply(lambda x : 1 if(x>3.5) else 0)
```

```python
df['stars'].unique()
```

```python
# Visualizing the Document Term Matrix using TF-IDF
import pandas as pd
vectors=pd.DataFrame(tfidf.toarray(), columns=vectorizer.get_feature_names())
#vectors['originalText']=pd.Series(df.text)
vectors['stars'] = pd.Series(df.stars)
vectors
```

```python
vectors = vectors.dropna()
vectors.stars.isna().sum()
```

```python
vectors.stars.unique()
```

```python
Y = vectors['stars']
Y.isnull().sum()
```

```python
del df
```

```python
vectors.drop(['stars'],axis =1,inplace = True)
```

```python
vectors.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X = pca.fit_transform(vectors)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

# Models

```python
del vectors
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC() # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, Y_train)
```

```python
#Predict the response for test dataset
predicted = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,predicted)
```
